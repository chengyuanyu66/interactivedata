```python
import pandas as pd
import re
from collections import Counter
import plotly.express as px

# 1️⃣ 读取数据
cexr = pd.read_csv("cexr.csv")

# 2️⃣ 标记是否属于 Linguistics 类别
cexr['is_linguistics'] = cexr['WoS Categories'].astype(str).str.contains(
    r"\blinguistics\b|\blanguage and linguistics\b", regex=True, case=False, na=False
)

# 3️⃣ 拆分关键词函数
def get_keywords(series):
    keywords = []
    for item in series.dropna():
        for kw in re.split(r'[;,]', str(item)):
            kw = kw.strip().lower()
            if kw:
                keywords.append(kw)
    return Counter(keywords)

# 4️⃣ 分别统计 Linguistics 与 Non-Linguistics
linguistics_kw = get_keywords(cexr.loc[cexr['is_linguistics'], 'Author Keywords'])
non_linguistics_kw = get_keywords(cexr.loc[~cexr['is_linguistics'], 'Author Keywords'])

# 5️⃣ 取前 20
linguistics_top20 = pd.DataFrame(linguistics_kw.most_common(20), columns=['Keyword', 'Frequency'])
non_linguistics_top20 = pd.DataFrame(non_linguistics_kw.most_common(20), columns=['Keyword', 'Frequency'])

# 6️⃣ 绘图 1：Linguistics 类别前 20
fig1 = px.bar(
    linguistics_top20,
    x='Keyword',
    y='Frequency',
    title='Top 20 Keywords — Linguistics Category',
    text='Frequency'
)
fig1.update_traces(textposition='outside')
fig1.update_layout(xaxis_tickangle=-45, height=600, margin=dict(l=60, r=20, t=80, b=120))
fig1.show()

# 7️⃣ 绘图 2：Non-Linguistics 类别前 20
fig2 = px.bar(
    non_linguistics_top20,
    x='Keyword',
    y='Frequency',
    title='Top 20 Keywords — Non-Linguistics Category',
    text='Frequency'
)
fig2.update_traces(textposition='outside')
fig2.update_layout(xaxis_tickangle=-45, height=600, margin=dict(l=60, r=20, t=80, b=120))
fig2.show()
```

```python
import pandas as pd
import re
import plotly.graph_objects as go

# ========== 1) 读取数据 ==========
df = pd.read_csv("cexr.csv")

# ========== 2) 标记是否属于 Linguistics 类别 ==========
df['is_linguistics'] = df['WoS Categories'].astype(str).str.contains(
    r"\blinguistics\b|\blanguage and linguistics\b",
    case=False, regex=True, na=False
)

# ========== 3) 使用 Times Cited, WoS Core 列 ==========
citation_col = "Times Cited, WoS Core"
df[citation_col] = pd.to_numeric(df[citation_col], errors="coerce").fillna(0).astype(int)

# ========== 4) 作者解析函数 ==========
def parse_last_names(authors_str: str):
    """按分号分作者，只取逗号前的姓（WoS 常见：Last, FirstInitials; Last, FirstInitials; ...）"""
    if pd.isna(authors_str) or not str(authors_str).strip():
        return []
    raw_authors = [a.strip() for a in str(authors_str).split(';') if a.strip()]
    last_names = []
    for a in raw_authors:
        if ',' in a:
            last_names.append(a.split(',')[0].strip())   # 逗号前是姓
        else:
            parts = a.split()
            last_names.append(parts[-1].strip())         # 无逗号时取最后一个词
    return last_names

def make_author_label(authors_str, year):
    """根据作者数量生成：Last, Year / Last & Last, Year / Last et al., Year"""
    last_names = parse_last_names(authors_str)
    y = str(year) if pd.notna(year) else ""
    if not last_names:
        return y
    if len(last_names) == 1:
        return f"{last_names[0]}, {y}"
    if len(last_names) == 2:
        return f"{last_names[0]} & {last_names[1]}, {y}"
    return f"{last_names[0]} et al., {y}"

# 生成横轴使用的 label
df['label'] = df.apply(
    lambda row: make_author_label(row.get('Authors', ''), row.get('Publication Year', '')),
    axis=1
)

# ========== 5) 各取前 20 篇高被引论文 ==========
ling_top = (
    df.loc[df['is_linguistics']]
      .sort_values(citation_col, ascending=False)
      .head(20)
)
nonling_top = (
    df.loc[~df['is_linguistics']]
      .sort_values(citation_col, ascending=False)
      .head(20)
)

# ========== 6) fig3：一张图，按钮切换 Linguistics / Non-Linguistics ==========
fig3 = go.Figure()

# Trace 0: Linguistics（默认可见）
fig3.add_bar(
    x=ling_top['label'],
    y=ling_top[citation_col],
    name="Linguistics",
    text=ling_top[citation_col],
    textposition="outside",
    hovertext=(
        "<b>Title:</b> " + ling_top['Article Title'].astype(str) +
        "<br><b>Journal:</b> " + ling_top['Source Title'].astype(str) +
        "<br><b>Year:</b> " + ling_top['Publication Year'].astype(str) +
        "<br><b>Cited:</b> " + ling_top[citation_col].astype(str)
    ),
    hoverinfo="text",
    visible=True
)

# Trace 1: Non-Linguistics（初始隐藏）
fig3.add_bar(
    x=nonling_top['label'],
    y=nonling_top[citation_col],
    name="Non-Linguistics",
    text=nonling_top[citation_col],
    textposition="outside",
    hovertext=(
        "<b>Title:</b> " + nonling_top['Article Title'].astype(str) +
        "<br><b>Journal:</b> " + nonling_top['Source Title'].astype(str) +
        "<br><b>Year:</b> " + nonling_top['Publication Year'].astype(str) +
        "<br><b>Cited:</b> " + nonling_top[citation_col].astype(str)
    ),
    hoverinfo="text",
    visible=False
)

# ========== 7) 按钮与布局（按钮与标题同一行靠右） ==========
buttons = [
    dict(
        label="Linguistics",
        method="update",
        args=[{"visible": [True, False]},
              {"title": "Top 20 Most Cited — Linguistics Category"}]
    ),
    dict(
        label="Non-Linguistics",
        method="update",
        args=[{"visible": [False, True]},
              {"title": "Top 20 Most Cited — Non-Linguistics Category"}]
    )
]

fig3.update_layout(
    updatemenus=[
        dict(
            type="buttons",
            direction="right",
            buttons=buttons,
            xanchor="left",
            yanchor="top",
            x=0.65,                 # 按钮水平位置（可调：0.55~0.85）
            y=1.08,                 # 按钮垂直位置（贴近标题）
            pad={"r": 8, "t": 4},
            showactive=True,
            bgcolor="rgba(0,0,0,0)",
            bordercolor="rgba(0,0,0,0)"
        )
    ],
    title=dict(
        text="Top 20 Most Cited — Linguistics Category",
        x=0.05,                    # 标题偏左
        xanchor="left",
        yanchor="top"
    ),
    xaxis=dict(tickangle=-45),
    yaxis_title="Times Cited (WoS Core)",
    height=700,
    margin=dict(l=80, r=40, t=100, b=220),
    showlegend=False
)

fig3.show()
```

```python
import pandas as pd
import re
import plotly.graph_objects as go

# =========================
# 可调参数（更开、更松）
TOP_N = 15
X_TICKANGLE = -60
X_TICKFONT_SIZE = 11
Y_TICKFONT_SIZE = 13
Y_TITLEFONT_SIZE = 15
FIG_HEIGHT = 1020
LEFT_MARGIN = 140          # ↑ 左侧多留白，给左侧图例腾空间的同时让y轴不挤
RIGHT_MARGIN = 40
TOP_MARGIN = 400           # ↑ 顶部更宽：播放键 + 时间轴 + 标题互不遮挡
BOTTOM_MARGIN = 360
BARGAP = 0.35
Y_MARGIN_RATIO = 1.25

# 顶部控件位置（按你的要求上移）
PLAYBTN_Y   = 1.70         # ↑ 播放/暂停更靠上
SLIDER_X    = 0.10         # 年份滑块略向中间
SLIDER_Y    = 1.45         # ↑ 年份滑块上移
SLIDER_LEN  = 0.86
PLAY_FRAME_MS = 800
PLAY_TRANS_MS = 200
# =========================

# 1) 读取数据
df = pd.read_csv("cexr.csv")

# 2) 标记 Linguistics 类别
df["is_linguistics"] = df["WoS Categories"].astype(str).str.contains(
    r"\blinguistics\b|\blanguage and linguistics\b", case=False, regex=True, na=False
)

# 3) 取需要的列并整理关键词
df_kw = df[["Publication Year", "Author Keywords", "is_linguistics"]].copy()
df_kw = df_kw.dropna(subset=["Author Keywords"])
df_kw["Publication Year"] = pd.to_numeric(df_kw["Publication Year"], errors="coerce")

def split_keywords(s: str):
    parts = re.split(r"[;,\|/]+", str(s))
    return [p.strip().lower() for p in parts if p.strip()]

df_kw["keyword"] = df_kw["Author Keywords"].apply(split_keywords)
df_kw = df_kw.explode("keyword")
df_kw = df_kw.dropna(subset=["keyword", "Publication Year"])

# 4) 统计每年、每组、每关键词频次
counts = (
    df_kw.groupby(["Publication Year", "is_linguistics", "keyword"])
         .size()
         .reset_index(name="freq")
)
counts["group"] = counts["is_linguistics"].map({True: "Linguistics", False: "Non-Linguistics"})

# 5) 为每年构建 Top-N 关键词并生成帧
years = sorted(counts["Publication Year"].dropna().unique())
frames = []
initial_year = int(years[0]) if len(years) > 0 else None

for yr in years:
    yri = int(yr)
    year_df = counts[counts["Publication Year"] == yr].copy()

    total_by_kw = year_df.groupby("keyword")["freq"].sum().sort_values(ascending=False)
    top_keywords = total_by_kw.head(TOP_N).index.tolist()
    year_top = year_df[year_df["keyword"].isin(top_keywords)].copy()

    pivot = year_top.pivot_table(
        index="keyword", columns="group", values="freq", aggfunc="sum", fill_value=0
    )
    for col in ["Linguistics", "Non-Linguistics"]:
        if col not in pivot.columns:
            pivot[col] = 0

    pivot["total"] = pivot["Linguistics"] + pivot["Non-Linguistics"]
    pivot = pivot.sort_values("total", ascending=False)

    x_keywords = pivot.index.tolist()
    y_ling = pivot["Linguistics"].clip(lower=0).tolist()
    y_non  = pivot["Non-Linguistics"].clip(lower=0).tolist()
    ymax = max([0] + y_ling + y_non) * Y_MARGIN_RATIO
    if ymax <= 0:
        ymax = 1.0

    frames.append(
        go.Frame(
            name=str(yri),
            data=[
                go.Bar(x=x_keywords, y=y_ling, name="Linguistics"),
                go.Bar(x=x_keywords, y=y_non,  name="Non-Linguistics")
            ],
            layout=go.Layout(
                title_text=f"Top {TOP_N} Keywords by Year — Linguistics vs Non-Linguistics ({yri})",
                xaxis=dict(
                    categoryorder="array",
                    categoryarray=x_keywords,
                    tickangle=X_TICKANGLE,
                    tickfont=dict(size=X_TICKFONT_SIZE),
                    ticklabelposition="outside bottom",
                    automargin=True
                ),
                yaxis=dict(
                    range=[0, ymax],
                    title="Frequency",
                    tickfont=dict(size=Y_TICKFONT_SIZE),
                    titlefont=dict(size=Y_TITLEFONT_SIZE),
                    automargin=True
                )
            )
        )
    )

# 6) 初始图
fig4 = go.Figure()
if initial_year is not None:
    init = [f for f in frames if f.name == str(initial_year)][0]
    for tr in init.data:
        fig4.add_trace(tr)
    title_text = init.layout.title.text
    init_x = init.data[0]["x"] if len(init.data) > 0 else []
    init_ymax = init.layout.yaxis.range[1]
else:
    fig4.add_trace(go.Bar(x=[], y=[], name="Linguistics"))
    fig4.add_trace(go.Bar(x=[], y=[], name="Non-Linguistics"))
    title_text = f"Top {TOP_N} Keywords by Year — Linguistics vs Non-Linguistics"
    init_x = []
    init_ymax = 10

fig4.frames = frames

# 7) 年份滑块（上部，略向中间）
steps = [
    dict(
        method="animate",
        args=[[str(int(yr))],
              dict(mode="immediate", frame=dict(duration=0, redraw=True),
                   transition=dict(duration=0))],
        label=str(int(yr))
    ) for yr in years
]

sliders = [dict(
    active=0,
    x=SLIDER_X, y=SLIDER_Y,
    len=SLIDER_LEN,
    xanchor="left",
    yanchor="top",
    pad={"t": 0, "b": 0},
    currentvalue={"prefix": "Year: ", "font": {"size": 16}},
    steps=steps
)]

# 8) 播放/暂停按钮（左上角，上移）
play_buttons = dict(
    type="buttons",
    direction="right",
    x=0.02, y=PLAYBTN_Y,
    xanchor="left",
    yanchor="top",
    pad={"r": 12, "t": 6},
    showactive=True,
    bgcolor="rgba(0,0,0,0)",
    bordercolor="rgba(0,0,0,0)",
    buttons=[
        dict(label="▶ Play",
             method="animate",
             args=[None, dict(fromcurrent=True,
                              frame=dict(duration=PLAY_FRAME_MS, redraw=True),
                              transition=dict(duration=PLAY_TRANS_MS))]),
        dict(label="⏸ Pause",
             method="animate",
             args=[[None], dict(mode="immediate",
                                frame=dict(duration=0, redraw=False),
                                transition=dict(duration=0))])
    ]
)

# 9) 布局：图例放到图外左侧，纵向；其余不变
fig4.update_layout(
    title=dict(text=title_text, x=0.02, xanchor="left"),
    xaxis=dict(
        tickangle=X_TICKANGLE,
        tickfont=dict(size=X_TICKFONT_SIZE),
        ticklabelposition="outside bottom",
        title="Keyword",
        categoryorder="array",
        categoryarray=init_x,
        automargin=True
    ),
    yaxis=dict(
        title="Frequency",
        tickfont=dict(size=Y_TICKFONT_SIZE),
        titlefont=dict(size=Y_TITLEFONT_SIZE),
        range=[0, init_ymax],
        automargin=True
    ),
    barmode="group",
    bargap=BARGAP,
    height=FIG_HEIGHT,
    margin=dict(l=LEFT_MARGIN, r=RIGHT_MARGIN, t=TOP_MARGIN, b=BOTTOM_MARGIN),
    legend=dict(
        orientation="v",
        x=-0.10, xanchor="right",   # ← 图外左边
        y=0.5,  yanchor="middle",
        bgcolor="rgba(255,255,255,0.9)",
        bordercolor="rgba(0,0,0,0.15)",
        borderwidth=1
    ),
    sliders=sliders,
    updatemenus=[play_buttons]
)

fig4.show()
```

```python
import pandas as pd
import re
import plotly.graph_objects as go

# ====== 显示&布局参数（与上图一致，保持风格统一） ======
TOP_N = 15
X_TICKANGLE = -60
X_TICKFONT_SIZE = 11
Y_TICKFONT_SIZE = 13
Y_TITLEFONT_SIZE = 15
FIG_HEIGHT = 1400
LEFT_MARGIN = 140
RIGHT_MARGIN = 40
TOP_MARGIN = 420
BOTTOM_MARGIN = 520
BARGAP = 0.35
Y_MARGIN_RATIO = 1.25
PLAYBTN_Y   = 1.72
SLIDER_X    = 0.10
SLIDER_Y    = 1.44
SLIDER_LEN  = 0.86
PLAY_FRAME_MS = 800
PLAY_TRANS_MS = 200
# ===============================================

# 1) 读数据
df = pd.read_csv("cexr.csv")

# 2) 标记 Linguistics 类别
df["is_linguistics"] = df["WoS Categories"].astype(str).str.contains(
    r"\blinguistics\b|\blanguage and linguistics\b", case=False, regex=True, na=False
)

# 3) 仅保留 Non-Linguistics 组，拆分关键词
sub = df[~df["is_linguistics"]].copy()
sub = sub[["Publication Year", "Author Keywords"]].dropna(subset=["Author Keywords"])
sub["Publication Year"] = pd.to_numeric(sub["Publication Year"], errors="coerce")

def split_keywords(s: str):
    return [p.strip().lower() for p in re.split(r"[;,\|/]+", str(s)) if p.strip()]

sub["keyword"] = sub["Author Keywords"].apply(split_keywords)
sub = sub.explode("keyword").dropna(subset=["keyword", "Publication Year"])

# 4) 每年每关键词频次
counts = sub.groupby(["Publication Year", "keyword"]).size().reset_index(name="freq")
years = sorted(counts["Publication Year"].dropna().unique())
frames = []
initial_year = int(years[0]) if len(years) else None

for yr in years:
    yri = int(yr)
    year_df = counts[counts["Publication Year"] == yr].sort_values("freq", ascending=False).head(TOP_N)
    x_keywords = year_df["keyword"].tolist()
    y_vals     = year_df["freq"].clip(lower=0).tolist()
    ymax = (max([0] + y_vals) * Y_MARGIN_RATIO) or 1.0

    frames.append(
        go.Frame(
            name=str(yri),
            data=[go.Bar(x=x_keywords, y=y_vals, name="Non-Linguistics")],
            layout=go.Layout(
                title_text=f"Non-Linguistics: Top {TOP_N} Keywords by Year ({yri})",
                xaxis=dict(categoryorder="array", categoryarray=x_keywords,
                           tickangle=X_TICKANGLE, tickfont=dict(size=X_TICKFONT_SIZE),
                           ticklabelposition="outside bottom", automargin=True),
                yaxis=dict(range=[0, ymax], title="Frequency",
                           tickfont=dict(size=Y_TICKFONT_SIZE), titlefont=dict(size=Y_TITLEFONT_SIZE),
                           automargin=True)
            )
        )
    )

# 5) 初始图
fig_nonling = go.Figure()
if initial_year is not None:
    init = [f for f in frames if f.name == str(initial_year)][0]
    for tr in init.data: fig_nonling.add_trace(tr)
    title_text = init.layout.title.text
    init_x = init.data[0]["x"] if len(init.data) > 0 else []
    init_ymax = init.layout.yaxis.range[1]
else:
    fig_nonling.add_trace(go.Bar(x=[], y=[], name="Non-Linguistics"))
    title_text, init_x, init_ymax = f"Non-Linguistics: Top {TOP_N} Keywords by Year", [], 10
fig_nonling.frames = frames

# 6) 年份滑块 + 播放键
steps = [dict(method="animate",
              args=[[str(int(y))], dict(mode="immediate", frame=dict(duration=0, redraw=True), transition=dict(duration=0))],
              label=str(int(y))) for y in years]
sliders = [dict(active=0, x=SLIDER_X, y=SLIDER_Y, len=SLIDER_LEN, xanchor="left", yanchor="top",
                pad={"t":0,"b":0}, currentvalue={"prefix":"Year: ","font":{"size":16}}, steps=steps)]
play_buttons = dict(type="buttons", direction="right", x=0.02, y=PLAYBTN_Y, xanchor="left", yanchor="top",
                    pad={"r":12,"t":6}, showactive=True, bgcolor="rgba(0,0,0,0)", bordercolor="rgba(0,0,0,0)",
                    buttons=[
                        dict(label="▶ Play",  method="animate",
                             args=[None, dict(fromcurrent=True, frame=dict(duration=PLAY_FRAME_MS, redraw=True),
                                              transition=dict(duration=PLAY_TRANS_MS))]),
                        dict(label="⏸ Pause", method="animate",
                             args=[[None], dict(mode="immediate", frame=dict(duration=0, redraw=False),
                                                transition=dict(duration=0))])
                    ])

# 7) 布局
fig_nonling.update_layout(
    title=dict(text=title_text, x=0.02, xanchor="left"),
    xaxis=dict(tickangle=X_TICKANGLE, tickfont=dict(size=X_TICKFONT_SIZE),
               ticklabelposition="outside bottom", title="Keyword",
               categoryorder="array", categoryarray=init_x, automargin=True),
    yaxis=dict(title="Frequency", tickfont=dict(size=Y_TICKFONT_SIZE),
               titlefont=dict(size=Y_TITLEFONT_SIZE), range=[0, init_ymax], automargin=True),
    showlegend=False, barmode="group", bargap=BARGAP,
    height=FIG_HEIGHT,
    margin=dict(l=LEFT_MARGIN, r=RIGHT_MARGIN, t=TOP_MARGIN, b=BOTTOM_MARGIN),
    sliders=sliders, updatemenus=[play_buttons]
)

fig_nonling.show()
# 可选：保存为HTML
# fig_nonling.write_html("fig_nonling.html", include_plotlyjs="cdn", full_html=True)
```
