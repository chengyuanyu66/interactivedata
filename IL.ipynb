```python
#Here I started to plot the most frequent journals.
frequency = il['Source Title'].value_counts().reset_index()
frequency.columns = ['Source Title', 'Frequency']
fig1 = px.bar(frequency, x='Source Title', y='Frequency', title='Frequency of Source Title')
fig1.show()
```

```python
#There are too many information in the bar plot above. It can be useful to zoom in.
# This is a top-20 solution.
frequency = il['Source Title'].value_counts().reset_index()
frequency.columns = ['Source Title', 'Frequency']
top_20_frequency = frequency.nlargest(20, 'Frequency')
fig2 = px.bar(top_20_frequency, x='Source Title', y='Frequency', title='Frequency of Top 20 Source Title')
fig2.show()
```

```python
#I would like to combine the low-frequency source titles.
other_frequency = frequency[~frequency['Source Title'].isin(top_20_frequency['Source Title'])]['Frequency'].sum()
others_df = pd.DataFrame({'Source Title': ['Others'], 'Frequency': [other_frequency]})
top_frequency_combined = pd.concat([top_20_frequency, others_df], ignore_index=True)
fig3 = px.bar(top_frequency_combined, x='Source Title', y='Frequency', title='Frequency of Top 20 Source Title (others)')
fig3.show()
# This figure looks weird. I may prefer fig2 to fig3.
```

```python
# This is a top-10 solution.
frequency = il['Source Title'].value_counts().reset_index()
frequency.columns = ['Source Title', 'Frequency']
top_10_frequency = frequency.nlargest(10, 'Frequency')
fig4 = px.bar(top_10_frequency, x='Source Title', y='Frequency', title='Frequency of Top 10 Source Title')
fig4.show()
```

```python
# This is a top-20 solution.
fig5 = px.pie(top_20_frequency, values='Frequency', names='Source Title', title='Top 20 Source Titles')
fig5.show()
```

```python
# This is a top-10 solution.
fig6 = px.pie(top_10_frequency, values='Frequency', names='Source Title', title='Top 10 Source Titles')
fig6.show()
```

```python
#Here I started to plot the most frequent authors.
frequency = il['Authors'].value_counts().reset_index()
frequency.columns = ['Authors', 'Frequency']
top_20_frequency = frequency.nlargest(20, 'Frequency')
fig10 = px.bar(top_20_frequency, x='Authors', y='Frequency', title='Top 20 Authors')
fig10.show()
# Some of the articles have multiple authors. It's better to parse them.
```

```python
il_authors_split = il['Authors'].str.split(';', expand=True)
il_authors_stacked = il_authors_split.stack().reset_index(drop=True)
author_frequency = il_authors_stacked.value_counts().reset_index()
author_frequency.columns = ['Authors', 'Frequency']
top_20_authors = author_frequency.nlargest(20, 'Frequency')
fig11 = px.bar(top_20_authors, x='Authors', y='Frequency', title='Top 20 Authors')
fig11.show()

# Why are there two bars for some scholars, for example, Julien H. They are the same, with no variations in form.
```

```python
fig17 = px.scatter(il,
                 x='Publication Year',
                 y='Times Cited, WoS Core',
                 height=500,
                 title="Times Cited, WoS Core",
                 hover_name="Article Title",
                 hover_data=["Authors","WoS Categories"],
                 )
fig17.show()
```

```python
threshold = 100

il_filtered = il[il['Times Cited, WoS Core'] >= threshold]

fig18 = px.scatter(il_filtered,
                   x='Publication Year',
                   y='Times Cited, WoS Core',
                   height=500,
                   title="Times Cited, WoS Core (Threshold: >= {})".format(threshold),
                   hover_name="Article Title",
                   hover_data=["Authors","WoS Categories"])

fig18.show()
```

```python
threshold = 100

fig19 = px.scatter(il_filtered,
                   x='Publication Year',
                   y='Times Cited, WoS Core',
                   title="Times Cited, WoS Core (Threshold: >= {})".format(threshold),
                   hover_name="Article Title",
                   hover_data=["Authors", "WoS Categories"],
                   color="WoS Categories")

fig19.show()
```

```python
threshold = 100

il_filtered = il[il['Times Cited, WoS Core'] >= threshold]

top_5_categories = il_filtered['WoS Categories'].value_counts().nlargest(5).index

il_filtered['WoS Categories Grouped'] = il_filtered['WoS Categories'].apply(lambda x: x if x in top_5_categories else 'Others')

fig20 = px.scatter(il_filtered,
                   x='Publication Year',
                   y='Times Cited, WoS Core',
                   title="Times Cited, WoS Core (Threshold: >= {})".format(threshold),
                   hover_name="Article Title",
                   hover_data=["Authors", "WoS Categories","Source Title"],
                   color="WoS Categories Grouped")  

fig20.show()
```
